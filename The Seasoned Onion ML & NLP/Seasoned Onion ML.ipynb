```python
import pandas as pd
import numpy as np

tags = pd.read_csv("tags_onehot_final3.csv")
tags = tags.drop("Unnamed: 0", axis=1)
tags
```

```python
body = pd.read_csv("body_onehot_final2.csv")
body = body.drop("Unnamed: 0", axis=1)
body
```

Here is our onehot encoded tags. This bag of words model will be our target values for our classifier. Before we move on, we have to merge this dataframe with our headlines data frame into one large model which we will feed into our classifier.

```python
#2856 is the column number where the last headline word is
data_onehot = pd.concat([body, tags], axis=1)
data_onehot
```

```python
def index_counter(array):

    index = -1 #Staring index is 0
    index_array = []
    for x in array:
        if x == 0:
            index = index + 1
        else:
            index = index +1
            index_array.append(index)
    return index_array
```

```python
index_array = index_counter(tag_array.flatten())
```

```python
tags_words_array = []
for i in index_array:  
    tags_words_array.append(y_test_df.columns.values[i])
```
